```python
column = df[['age','campaign','duration']]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 - 1.5 * iqr
df[['age','campaign','duration']] = column[(column > lower_bound) & (column < upper_bound)]
```

```python
from sklearn.preprocessing import LabelEncoder # Fixed the case of LabelEncoder
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded # Fixed the variable name here
```

```python
df_encoded['deposit'].value_counts()
```

```python
x = df_encoded.drop('deposit',axis=1)
y = df_encoded['deposit']
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score

def eval_model(y_test,y_pred):
  acc = accuracy_score(y_test,y_pred)
  print('Accuracy_Score',acc)
  cm = confusion_matrix(y_test,y_pred)
  print('Confusion Matrix\n',cm)
  print('Classification Report\n',classification_report(y_test,y_pred))

# Fix indentation for mscore function
def mscore(model):
  # Assuming you meant to use x_train and x_test here
  train_score = model.score(x_train,y_train)
  test_score = model.score(x_test,y_test)
  print('Training Score',train_score)
  print('Testing Score',test_score)
```

```python
from sklearn.tree import DecisionTreeClassifier # Fix the typo

dt = DecisionTreeClassifier(criterion='gini',max_depth=5,min_samples_split=10)
dt.fit(x_train,y_train) # Fit the decision tree model
```

```python
mscore(dt)
```

```python
y_pred_dt = dt.predict(x_test)
print(y_pred_dt) # Use the correct variable name y_pred_dt
```

```python
eval_model(y_test,y_pred_dt) # Fix the typo in the variable name
```
